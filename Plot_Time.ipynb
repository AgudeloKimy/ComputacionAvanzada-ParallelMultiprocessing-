```python
numbers={"Núcleos": [ 1,  2, 3, 4, 5, 6, 7, 8],  "Tiempo": [9.17911529541,  8.559226989, 8.5353851318, 6.17504119873, 6.031990051269, 5.6743621826171, 5.626678466796, 5.6266784667968]  }  #  Second key-lis
```
